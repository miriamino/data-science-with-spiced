```python
import pandas as pd
```

```python
df = pd.read_csv('gapminder_total_fertility.csv', index_col=0)
df
```

```python
df = df[['1916', '1977', '2013']]
```

```python
df = df.loc[['Bangladesh', 'Brazil', 'China', ], :]
df
```

```python
from matplotlib import pyplot as plt
df.plot.bar(figsize=(5,7))
plt.savefig('bars.png')
```

```python
df.plot.bar(y='1916', width=0.7, color='orange')
```

```python
df = df.transpose()
df.plot.bar(figsize=(5,7))
plt.savefig('barstransposed.png')
```
