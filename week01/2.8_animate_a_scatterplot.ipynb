#### Step 6
Now convert the table into long format. First, we have to move the row index into a column:

```python
fert = fert.reset_index()
fert
```

```python
life = life.reset_index()
life
```

Second, we use the convenient melt function for the conversion:

```python
fert = fert.melt(id_vars='country', var_name='year', value_name='fertility_rate')
fert
```

```python
life = life.melt(id_vars='country', var_name='year', value_name='life_expectancy')
life
```

#### Step 7
Repeat the steps above to add data from the file data/gapminder_population.xlsx.

```python
population = pd.read_excel('data/gapminder_population.xlsx', index_col=0)
population.columns
population.index.name = 'country'
population = population.reset_index()
population
```

```python
population = population.melt(id_vars='country', var_name='year', value_name='population')
population
```

#### Step 8
Now we can glue the data frames together. This works great with the merge function:

```python
df = fert.merge(population)
df = df.merge(life)
df
```

```python
co2 = pd.read_csv('data/yearly_co2_emissions_1000_tonnes.csv', index_col=0)
co2.columns = co2.columns.astype(int)
co2 = co2.reset_index()
co2 = co2.melt(id_vars='country', var_name='year', value_name='co2_emissions')
df = df.merge(co2)
df
```

```python
cumco2 = pd.read_csv('data/cumulative_co2_emissions_tonnes.csv', index_col=0)
cumco2.columns = cumco2.columns.astype(int)
cumco2 = cumco2.reset_index()
cumco2 = cumco2.melt(id_vars='country', var_name='year', value_name='cum_co2_emissions')
df = df.merge(cumco2)
```

#### Step 12
Install the Python module imageio, by typing in the terminal:

Adjust the following code example and execute it:

```python
import imageio
import os

images = []

folder = 'images'

if not os.path.exists(folder):
    os.mkdir(folder)
```

```python
pop_min = df['population'].min()
pop_max = df['population'].max()
pop_min

co2_min = df['cum_co2_emissions'].min()
co2_max = df['cum_co2_emissions'].max()/2
```

```python
for year in range(1960, 2016):
    p = sns.scatterplot(x='life_expectancy', 
                        y='fertility_rate', 
                        hue='cum_co2_emissions', 
                        data=df[df['year'] == year], 
                        alpha=0.6, size='population', 
                        legend=False, sizes=(50, 4000), 
                        size_norm=(pop_min, pop_max),
                        hue_norm=(co2_min, co2_max))
    plt.title(f'Year: {year}')
    p.set(xlabel='Life Expectancy', ylabel='Total Fertility Rate')
    plt.axis([25, 80, 1, 9])
    filename = f'lifeexpfert_{year}.png'
    plt.savefig(os.path.join(folder,filename))
    images.append(imageio.imread(os.path.join(folder,filename)))
    plt.figure()    
imageio.mimsave(os.path.join(folder,'output.gif'), images, fps=20)
```
